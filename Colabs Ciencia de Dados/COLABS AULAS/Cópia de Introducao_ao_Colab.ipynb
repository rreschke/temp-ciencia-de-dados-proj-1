```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Cal2Joule(val):
  return val*4.184
```

```python
df = pd.DataFrame()
df['cal'] = np.arange(-20, 20)
df['joule'] = Cal2Joule(df['cal'])
df.head()
```

```python
from sklearn.tree import DecisionTreeRegressor


X = df[['cal']] 
y = df['joule']

regressor = DecisionTreeRegressor()
regressor.fit(X, y)
```

```python
joule_prediction = regressor.predict(pd.DataFrame({'cal':[0,27]}))
print(joule_prediction)
```

```python
# The coefficients
print("Coefficients: \n", regressor.coef_, regressor.intercept_)
```
