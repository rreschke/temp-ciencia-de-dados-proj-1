```python
from sklearn.linear_model import LogisticRegression

# Entradas e Saídas
X = breast.drop(columns=['id','diagnosis'])
y = breast['diagnosis']

# Definição
clf = LogisticRegression(max_iter=10000)

# Treinamento
clf.fit(X,y)
```

## Regressão Logística $\times$ Regressão Linear (RESOLVIDO)

Substitua os valores de `diagnosis` por `'B':1,'M':0` e converta para um tipo numérico (`float`, por exemplo). Os valores agora são numéricos e você pode, então, aplicar uma regressão linear para estimar uma função linear $f(X)$, com  $y=0$, se $f(X) > 0.5$ e $y=1$, se $f(X) \le 0.5$.

>> **Embora neste caso os resultados sejam os mesmos, a regressão linear apresenta em geral um desempenho bastante pior quando aplicada em problemas de classificação, em especial para as classes de valores próximos.**

```python
breast['diagnosis'] = breast['diagnosis'].replace({'B':1,'M':0})
breast['diagnosis'] = breast['diagnosis'].astype('float')

breast['diagnosis']
```

```python
from sklearn.linear_model import LogisticRegression

# Entradas e Saídas
X = breast.drop(columns=['id','diagnosis'])
X = X / X.max()
y = breast['diagnosis']

# Definição
clf = LogisticRegression(max_iter=10000)

# Treinamento
clf.fit(X,y)

print(clf.score(X,y))
```

```python
from sklearn.linear_model import LinearRegression

# Entradas e Saídas
X = breast.drop(columns=['id','diagnosis'])
X = X / X.max()
y = breast['diagnosis']

# Definição
clf = LinearRegression()

# Treinamento
clf.fit(X,y)

print(clf.score(X,y))

# Avaliação
y_pred = clf.predict(X) > 0.5

# Acuracidade
print(sum(y_pred == y) / len(y))
```
