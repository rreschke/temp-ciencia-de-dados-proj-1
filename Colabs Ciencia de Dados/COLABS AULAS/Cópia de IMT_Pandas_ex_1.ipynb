## **CASE Tips** 

##### Q1. Considere a base de dados [`tips`](https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/tips.csv). Considerando somente os finais de semana, quem tem média de participantes na mesa maior, fumantes ou não fumantes?



```python
tips = pd.read_csv('https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/tips.csv')
tips['smoker'].value_counts()
```

```python
df_naofumantes_tips = tips[tips['smoker'] == 'No']
df_fumantes_tips = tips[tips['smoker'] == 'Yes']

participantes_nf = df_naofumantes_tips.sum()['size']
participantes_f = df_naofumantes_tips.sum()['size']

mnf = participantes_nf/len(df_naofumantes_tips)
mf = participantes_f/len(df_fumantes_tips)

print(mf>mnf)
```

## **CASE Insurance** 

Acesse a base de dados [`insurance`](https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/insurance.csv) para as questões a seguir.



```python
insur = pd.read_csv('https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/insurance.csv')
insur
```

##### Q2. Qual percentual de homens e mulheres assegurados na base? 

```python
m = len(insur[insur['sex'] == 'male'])
f = m = len(insur[insur['sex'] == 'female'])

mpc = m / (m+f)
fpc = f / (m+f)

print(f'mpc = {mpc}; fpc = {fpc}')
```

##### Q3. Qual o total do valor de seguro pago por homens e mulheres segurados na base? 

```python
insur.groupby('sex').sum()
```

##### Q4. Qual o precentual de valor de seguros pagos por homens e mulheres segurados na base? 

```python
m = insur[insur['sex'] == 'male']['charges'].sum()
f = insur[insur['sex'] == 'female']['charges'].sum()

mpc = m / (m+f)
fpc = f / (m+f)
print(f'mpc = {mpc}; fpc = {fpc}')
```

##### Q5. Não fumantes apresentam em média mais filhos? 

```python
df_naofumantes = insur[insur['smoker'] == 'no']
df_fumantes = insur[insur['smoker'] == 'yes']
cnf = df_naofumantes['children'].sum()
print(f'Filhos de não fumantes: {cnf}')
cf = df_fumantes['children'].sum()
print(f'Filhos de fumantes: {cf}')
qnt_naofumantes = len(df_naofumantes)
qnt_fumantes = len(df_fumantes)

mcnf = cnf / qnt_naofumantes
mcf = cf / qnt_fumantes

print(f'Média de filhos por não fumantes: {mcnf}\nMédia por fumantes: {mcf}')
print(mcnf>mcf)
```

```python
energy = pd.read_csv('https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/energy_types.csv')
energy.head()
```

##### Q7. Quantos tipos de energia e países diferentes há na base?

```python
types = len(energy['type'].unique())
paises = len(energy['country'].unique())

print(f'Tipos: {types}\nPaíses: {paises}')
```

##### Q8. Qual o percentual de energia nuclear produzido na Europa no ano de 2018?

```python
df_nuclear_2018 = energy[energy['type'] == 'Nuclear'][['country','type','2018']]
df_nuclear_2018.head()
```

```python
df_2018 = energy[['country','type','2018']]
df_2018.head()
```

```python
pc = df_nuclear_2018.sum()['2018']/df_2018.sum()['2018']
print(f'Percentual de energia nuclear na Europa em 2018: {pc}')
```

##### Q9. Qual o percentual de energia nuclear produzido na França no ano de 2018?

```python
df_nuclear_fr_2018 = df_nuclear_2018[df_nuclear_2018['country'] == 'FR']
df_nuclear_fr_2018
```

```python
df_fr_2018 = df_2018[df_2018['country'] == 'FR']
df_fr_2018.head()
```

```python
pc_nuclear_fr_2018 = df_nuclear_fr_2018.sum()['2018']/df_fr_2018.sum()['2018']
print(f'Percentual de energia nuclear na França em 2018: {pc_nuclear_fr_2018}')
```

##### Q10. Qual o percentual de energias limpas produzida na Europa no ano de 2018?

*Energias limpas somente Hidroelétrica, Eólica, Solar e Geotérmica.*

```python
pd.DataFrame(energy['type'].unique())
```

```python
limpo = ['Hydro', 'Geothermal', 'Wind', 'Solar']
df_limpo_2018 = df_2018[df_2018['type'].isin(limpo)]
df_limpo_2018.head()
```

```python
en_limpo_2018 = df_limpo_2018.sum()['2018']
en_total_2018 = df_2018.sum()['2018']

pc_limpo_2018 = en_limpo_2018/en_total_2018

print(f'Percentual de energia limpa na Europa em 2018: {pc_limpo_2018}')
```

##### Q11. Considerando os anos de 2016, 2017 e 2018, a produção de energias limpas tem aumentado na Europa? Verifique em termos absolutos e percentuais.

*Energias limpas somente Hidroelétrica, Eólica, Solar e Geotérmica.*

```python
df_limpo = energy[energy['type'].isin(limpo)]
df_limpo.head()
```

```python
df_sujo = energy[~energy['type'].isin(limpo)]
df_sujo.head()
```

```python
en_limpa = df_limpo.sum(numeric_only=True)
en_suja = df_sujo.sum(numeric_only=True)

en_limpa_2016 = en_limpa['2016']
en_limpa_2017 = en_limpa['2017']
en_limpa_2018 = en_limpa['2018']

en_sujo_2016 = en_suja['2016']
en_sujo_2017 = en_suja['2017']
en_sujo_2018 = en_suja['2018']

print(f'Produção de energia limpa na Europa:\n2016: {en_limpa_2016}\n2017: {en_limpa_2017}\n2018: {en_limpa_2018}\n')
print(f'Produção de energia suja na Europa:\n2016: {en_sujo_2016}\n2017: {en_sujo_2017}\n2018: {en_sujo_2018}\n')

print(f'Percentual limpo:\n2016: {en_limpa_2016/en_sujo_2016}\n2017: {en_limpa_2017/en_sujo_2017}\n2018: {en_limpa_2018/en_sujo_2018}')
```
