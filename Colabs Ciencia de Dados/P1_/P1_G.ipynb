````python
from sklearn.metrics import confusion_matrix, classification_report

def calcular_metricas_mc(confusion_matrix, classe):
    # Extrair as classes do DataFrame
    classes = confusion_matrix.columns.tolist()

    # Verificar se a classe fornecida está presente na matriz de confusão
    if classe not in classes:
        print("A classe fornecida não está presente na matriz de confusão.")
        return


    # Calcular os valores relevantes da matriz de confusão
    true_positive = confusion_matrix.loc[classe, classe]
    false_positive = confusion_matrix[classes].sum()[classe] - true_positive
    false_negative = confusion_matrix.loc[classe, classes].sum() - true_positive

    # Calcular as métricas

    # Precisão = TP / (TP + FP) - coluna
    precision = true_positive / (true_positive + false_positive)

    # Recall = TP / (TP + FN) - linha
    recall = true_positive / (true_positive + false_negative)

    # F1 = 2 * (Precisão * Recall) / (Precisão + Recall)
    f1_score = 2 * ((precision * recall) / (precision + recall))


    # Imprimir as métricas
    print(f"A precisão da classe {classe} é {precision:.2f}")
    print(f"O recall da classe {classe} é {recall:.2f}")
    print(f"O F1 da classe {classe} é {f1_score:.2f}")


y_pred = classificador.predict(X_test)

print('Matriz de confusão:')
cm = confusion_matrix(y_test, y_pred)

cm_df = pd.DataFrame(cm,index=[classificador.classes_],columns=[classificador.classes_])
cm_df.columns = classificador.classes_
cm_df.index = classificador.classes_
display(cm_df)
print()


# Classes
classes = classificador.classes_
print(f'Classes: {classes}')
print()

for classe in classes:
    calcular_metricas_mc(cm_df, classe)
    print()
````

Recall (Versicolor): 1.00 = Todos os casos versicolor classificados corretamente, 25/25
